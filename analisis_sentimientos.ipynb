```python
#Importamos el paquete
from sentiment_analysis_spanish import sentiment_analysis
```

```python
#Importamos la clase necesaria para el análisis y creamos el objeto llamado sentiment para utilizar las funciones de la clase
sentiment = sentiment_analysis.SentimentAnalysisSpanish()
```

Ejecutaremos 3 ejemplos: un mensaje positivo, un mensaje negativo y otro neutral. Esto para comprobar que la librería detecta la polaridad de cada uno de estos mensajes

```python
x = "Esta muy buena la pelicual"
y = "Que horrible la comida!!!"
z = "Tuve una experiencia neutral"
```

```python
resultado1 = sentiment.sentiment(x)
print(resultado1)
```

```python
resultado2 = sentiment.sentiment(y)
print(resultado2)
```

```python
resultado3 = sentiment.sentiment(z)
print(resultado3)
```
